# Combination of all the necessary data

## Class for all types of dwellings

```python
class All_Types:
    def __init__(self, *args):
        self.df_1 = df_Climate_Zone
        self.df_2 = df_Housing_Characteristics
        self.df_3 = df_Headcount
        self.df_4 = df_Heating_System
        self.df_5 = df_Economic_Hypothesis
        self.df_6 = df_Wall_Investment
        self.df_7 = df_Window_Investment
        self.df_8 = df_Roof_Investment
        self.df_9 = df_Floor_Investment
```

## Class for CID type dwellings

```python
class CID_Type(All_Types):
    def init(self, *args):
        super().init(*args)

    def Combine_Lists(self):
        self.list_combined = []
        list_dwelling_type = ['CID']
        list_climate_type = self.df_1.iloc[:, 0].tolist()
        list_construction_period = self.df_2.iloc[:21, 1].tolist()[::3]
        list_size_factor = self.df_2.iloc[:3, 2].tolist()
        list_heating_system = [
            'ER', 'GB', 'CGB', 'OB', 'COB', 'EHP2', 'WB', 'WPB', 'EHP'
        ]
        list_initial_age = self.df_3.iloc[:20, 5].tolist()
        self.list_combined.append(list_dwelling_type)
        self.list_combined.append(list_climate_type)
        self.list_combined.append(list_construction_period)
        self.list_combined.append(list_size_factor)
        self.list_combined.append(list_heating_system)
        self.list_combined.append(list_initial_age)
        return self.list_combined

    def Generate_DF(self):
        self.Combine_Lists()
        itertools_product = list(itertools.product(*self.list_combined))
        self.df = pd.DataFrame(itertools_product,
                               columns=[
                                   'Dwelling Type', 'Climate Type',
                                   'Construction Period', 'Size Factor',
                                   'Heating System', 'Initial Age'
                               ])
        return self.df

    def Merge_DFs(self):
        self.Generate_DF()
        self.df = self.df.merge(self.df_1, how='left', on=['Climate Type'])
        self.df = self.df.merge(self.df_2.iloc[:21, 1:9],
                                how='left',
                                on=['Construction Period', 'Size Factor'])
        self.df = self.df.merge(self.df_4.iloc[:9, :],
                                how='left',
                                on=['Heating System'])
        return self.df

    def Add_Columns(self):
        self.Merge_DFs()
        self.df[self.df_5.columns[0]] = self.df_5.iloc[0, 0]
        self.df[self.df_5.columns[1]] = self.df_5.iloc[0, 1]
        self.df[self.df_5.columns[2]] = self.df_5.iloc[0, 2]
        self.df[self.df_5.columns[3]] = self.df_5.iloc[0, 3]
        return self.df

    def Concat_DFs(self):
        self.Add_Columns()
        self.df = pd.concat([self.df, self.df_3.iloc[0:11340, 6]],
                            axis=1,
                            join='outer')
        return self.df

    def Reorder_Columns(self):
        self.Concat_DFs()
        cols = self.df.columns.tolist()
        reordered_cols = cols[:2] + cols[6:7] + cols[2:4] + cols[7:13] + \
        cols[4:5] + cols[13:27] + cols[5:6] + cols[27:28]
        self.df = self.df[reordered_cols]
        return self.df

    def Add_U_Factors(self):
        self.Reorder_Columns()
        self.df = self.df.merge(
            self.df_2.loc[:20, ['Construction Period', 'Size Factor', 'Um0']],
            how='left',
            on=['Construction Period', 'Size Factor'])
        for m in range(18):
            self.df['Um' + str(m + 1)] = self.df_6.iloc[m, 2]
        self.df = self.df.merge(
            self.df_2.loc[:20, ['Construction Period', 'Size Factor', 'Uw0']],
            how='left',
            on=['Construction Period', 'Size Factor'])
        for w in range(3):
            self.df['Uw' + str(w + 1)] = self.df_7.iloc[w, 2]
        self.df = self.df.merge(
            self.df_2.loc[:20, ['Construction Period', 'Size Factor', 'Ur0']],
            how='left',
            on=['Construction Period', 'Size Factor'])
        for r in range(3):
            self.df['Ur' + str(r + 1)] = self.df_8.iloc[r, 2]
        self.df = self.df.merge(
            self.df_2.loc[:20, ['Construction Period', 'Size Factor', 'Uf0']],
            how='left',
            on=['Construction Period', 'Size Factor'])
        for f in range(5):
            self.df['Uf' + str(f + 1)] = self.df_9.iloc[f, 2]
        return self.df

    def Add_U_Prices(self):
        self.Reorder_Columns()
        self.df['Um0'] = 0
        for m in range(18):
            self.df['Um' + str(m + 1)] = self.df_6.iloc[m, 3]

        self.df['Uw0'] = 0
        for w in range(3):
            self.df['Uw' + str(w + 1)] = self.df_7.iloc[w, 3]

        self.df['Ur0'] = 0
        for r in range(3):
            self.df['Ur' + str(r + 1)] = self.df_8.iloc[r, 3]

        self.df['Uf0'] = 0
        for f in range(5):
            self.df['Uf' + str(f + 1)] = self.df_9.iloc[f, 3]
        return self.df
```

## Class for the collective dwellings

---

* **Formulas related to the surface component**:

\begin{align}
&\color{purple}{B_{surface\_m}} = \left(1-\color{teal}{D_{share\_of\_window}}\right) \times \sqrt{\color{teal}{D_{surface}} \times \color{teal}{B_{dwelling\_per\_floor}}} \times 4 \times \color{teal}{D_{ceiling\_height}} \times \left(\color{teal}{B_{floors}} + 1\right) \\
\notag \\
&\color{purple}{B_{surface\_w}} = \color{teal}{D_{share\_of\_window}} \times \sqrt{\color{teal}{D_{surface}} \times \color{teal}{B_{dwelling\_per\_floor}}} \times 4 \times \color{teal}{D_{ceiling\_height}} \times \left(\color{teal}{B_{floors}} + 1\right) \\
\notag \\
&\color{purple}{B_{surface\_r}} = \color{teal}{D_{surface}} \times \color{teal}{B_{dwelling\_per\_floor}} \\
\notag \\
&\color{purple}{B_{dwelling\_total}} = \color{teal}{B_{dwelling\_per\_floor}} \times \left(\color{teal}{B_{floors}} + 1\right)
\end{align}

* **Formulas related to the overall heat transfer coefficient ($Ug$)**:

\begin{align}
&\color{brown}{B_{weighted\_surface\_total}} = \color{red}{B_{weighted\_surface\_m\_w}} + \color{red}{B_{weighted\_surface\_r\_f}} \\
\notag \\
&\color{red}{B_{weighted\_surface\_m\_w}} = \color{purple}{B_{surface\_m}} \times \color{red}{D_{um}} + \color{purple}{B_{surface\_w}} \times \color{red}{D_{uw}} \\
\notag \\
&\color{red}{B_{weighted\_surface\_r\_f}} = \color{purple}{B_{surface\_r}} \times \left(\color{red}{D_{ur}} + \color{red}{D_{uf}}\right)
\end{align}

---

```python
class Calculable_Columns_for_CD(Calculable_Columns):
    def init(self, df):
        super().init(df)

    def Rename_Columns(self):
        self.df.rename(columns={
            name: name.replace(' ', '_')
            for name in self.df.columns if name.count(' ') > 0
        },
                       inplace=1)
        return self.df

    def Surface_M(self):
        self.Rename_Columns()
        A = np.subtract(1, self.df.Share_of_Windows)
        B = np.sqrt(np.multiply(self.df.Surface, self.df.Dwelling_per_Floor))
        A = np.multiply(np.multiply(A, B), 4)
        B = np.add(self.df.Floors, 1)
        B = np.multiply(self.df.Ceiling_Height, B)
        self.df['Surface_M'] = np.multiply(A, B)
        del A, B
        return self.df

    def Surface_W(self):
        self.Rename_Columns()
        A = np.sqrt(np.multiply(self.df.Surface, self.df.Dwelling_per_Floor))
        A = np.multiply(np.multiply(self.df.Share_of_Windows, A), 4)
        B = np.add(self.df.Floors, 1)
        B = np.multiply(self.df.Ceiling_Height, B)
        self.df['Surface_W'] = np.multiply(A, B)
        del A, B
        return self.df

    def Surface_R(self):
        self.Rename_Columns()
        self.df['Surface_R'] = np.multiply(self.df.Surface,
                                           self.df.Dwelling_per_Floor)
        return self.df

    def Weight_U_Factors_1(self):
        self.Surface_M()
        self.Surface_W()
        self.Surface_R()
        self.df.loc[:, 'Um0':'Um18'] = self.df.loc[:, 'Um0':'Um18'].apply(
            lambda x: np.multiply(x, self.df.Surface_M))
        self.df.loc[:, 'Uw0':'Uw3'] = self.df.loc[:, 'Uw0':'Uw3'].apply(
            lambda x: np.multiply(x, self.df.Surface_W))
        self.df.loc[:, 'Ur0':'Ur3'] = self.df.loc[:, 'Ur0':'Ur3'].apply(
            lambda x: np.multiply(x, self.df.Surface_R))
        self.df.loc[:, 'Uf0':'Uf5'] = self.df.loc[:, 'Uf0':'Uf5'].apply(
            lambda x: np.multiply(x, self.df.Surface_R))
        return self.df

    def Weight_U_Factors_2(self):
        self.Weight_U_Factors_1()
        columns_list = []
        columns_list_1 = []
        columns_list_2 = []
        for m in range(19):
            for w in range(4):
                column_name = 'U_m' + str(m) + '_w' + str(w)
                columns_list_1.append(column_name)
        for r in range(4):
            for f in range(6):
                column_name = 'U_r' + str(r) + '_f' + str(f)
                columns_list_2.append(column_name)
        columns_list = columns_list_1 + columns_list_2
        self.df = pd.concat([
            self.df,
            pd.DataFrame(0, index=self.df.index, columns=columns_list)
        ],
                            axis=1)
        self.df[columns_list_1] = self.df[columns_list_1].apply(
            lambda x: np.add(self.df['U' + x.name.split('_')[1]], self.df[
                'U' + x.name.split('_')[2]]))
        self.df[columns_list_2] = self.df[columns_list_2].apply(
            lambda x: np.add(self.df['U' + x.name.split('_')[1]], self.df[
                'U' + x.name.split('_')[2]]))
        return self.df

    def Dwelling_Total(self):
        self.Rename_Columns()
        self.df['Dwelling_Total'] = np.multiply(self.df.Dwelling_per_Floor,
                                                np.add(self.df.Floors, 1))
        return self.df
```

```python
class Calculable_DFs:
    def __init__(self, df):
        self.df = df
```

```python
class Calculable_DFs_for_CD(Calculable_DFs):
    def init(self, df):
        super().init(df)

    def Add_ER_Columns(self):
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'ER_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df = pd.concat([
            self.df,
            pd.DataFrame(0, index=self.df.index, columns=columns_list)
        ],
                            axis=1)
        return self.df

    def Fill_ER_Columns(self):
        self.Add_ER_Columns()
        selected_columns = self.df.loc[:, 'ER_m0_w0_r0_f0':].columns.tolist()
        self.df[selected_columns] = self.df[selected_columns].apply(
            lambda x: np.add(
                self.df['U_' + x.name.split('_')[1] + '_' + x.name.split('_')[
                    2]], \
                self.df['U_' + x.name.split('_')[3] + '_' + x.name.
                                 split('_')[4]]))
        return self.df

    def Generate_GB_Columns(self):
        self.Fill_ER_Columns()
        self.df_GB = self.df.loc[:, 'ER_m0_w0_r0_f0':'ER_m18_w3_r3_f5']
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'GB_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df_GB.columns = columns_list
        self.df = pd.concat([self.df, self.df_GB], axis=1)
        del self.df_GB
        return self.df

    def Generate_CGB_Columns(self):
        self.Generate_GB_Columns()
        self.df_CGB = self.df.loc[:, 'ER_m0_w0_r0_f0':'ER_m18_w3_r3_f5']
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'CGB_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df_CGB.columns = columns_list
        self.df = pd.concat([self.df, self.df_CGB], axis=1)
        del self.df_CGB
        return self.df

    def Generate_OB_Columns(self):
        self.Generate_CGB_Columns()
        self.df_OB = self.df.loc[:, 'ER_m0_w0_r0_f0':'ER_m18_w3_r3_f5']
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'OB_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df_OB.columns = columns_list
        self.df = pd.concat([self.df, self.df_OB], axis=1)
        del self.df_OB
        return self.df

    def Generate_COB_Columns(self):
        self.Generate_OB_Columns()
        self.df_COB = self.df.loc[:, 'ER_m0_w0_r0_f0':'ER_m18_w3_r3_f5']
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'COB_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df_COB.columns = columns_list
        self.df = pd.concat([self.df, self.df_COB], axis=1)
        del self.df_COB
        return self.df

    def Generate_EHP2_Columns(self):
        self.Generate_COB_Columns()
        self.df_EHP2 = self.df.loc[:, 'ER_m0_w0_r0_f0':'ER_m18_w3_r3_f5']
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'EHP2_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df_EHP2.columns = columns_list
        self.df = pd.concat([self.df, self.df_EHP2], axis=1)
        del self.df_EHP2
        return self.df

    def Generate_WB_Columns(self):
        self.Generate_EHP2_Columns()
        self.df_WB = self.df.loc[:, 'ER_m0_w0_r0_f0':'ER_m18_w3_r3_f5']
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'WB_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df_WB.columns = columns_list
        self.df = pd.concat([self.df, self.df_WB], axis=1)
        del self.df_WB
        return self.df

    def Generate_WPB_Columns(self):
        self.Generate_WB_Columns()
        self.df_WPB = self.df.loc[:, 'ER_m0_w0_r0_f0':'ER_m18_w3_r3_f5']
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'WPB_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df_WPB.columns = columns_list
        self.df = pd.concat([self.df, self.df_WPB], axis=1)
        del self.df_WPB
        return self.df

    def Generate_EHP_Columns(self):
        self.Generate_WPB_Columns()
        columns_list = []
        self.df_EHP = self.df.loc[:, 'ER_m0_w0_r0_f0':'ER_m18_w3_r3_f5']
        columns_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        column_name = 'EHP_m' + str(m) + '_w' + str(
                            w) + '_r' + str(r) + '_f' + str(f)
                        columns_list.append(column_name)
        self.df_EHP.columns = columns_list
        self.df = pd.concat([self.df, self.df_EHP], axis=1)
        del self.df_EHP
        return self.df

    ####
    def Reproduce_ER_Columns(self):
        self.Fill_ER_Columns()
        columns_list = []
        for sh in ('GB CGB OB COB EHP2 WB WPB EHP').split():
            for m in range(19):
                for w in range(4):
                    for r in range(4):
                        for f in range(6):
                            column_name = sh + '_m' + str(m) + '_w' + str(
                                w) + '_r' + str(r) + '_f' + str(f)
                            columns_list.append(column_name)
        self.df = pd.concat([
            self.df,
            pd.DataFrame(0, index=self.df.index, columns=columns_list)
        ],
                            axis=1)
        copied_columns = self.df.loc[:, 'ER_m0_w0_r0_f0':
                                     'ER_m18_w3_r3_f5'].columns.tolist()
        selected_columns = self.df.loc[:, 'GB_m0_w0_r0_f0':
                                       'GB_m18_w3_r3_f5'].columns.tolist()
        self.df[selected_columns] = self.df[copied_columns]
        selected_columns = self.df.loc[:, 'CGB_m0_w0_r0_f0':
                                       'CGB_m18_w3_r3_f5'].columns.tolist()
        self.df[selected_columns] = self.df[copied_columns]
        selected_columns = self.df.loc[:, 'OB_m0_w0_r0_f0':
                                       'OB_m18_w3_r3_f5'].columns.tolist()
        self.df[selected_columns] = self.df[copied_columns]
        selected_columns = self.df.loc[:, 'COB_m0_w0_r0_f0':
                                       'COB_m18_w3_r3_f5'].columns.tolist()
        self.df[selected_columns] = self.df[copied_columns]
        selected_columns = self.df.loc[:, 'EHP2_m0_w0_r0_f0':
                                       'EHP2_m18_w3_r3_f5'].columns.tolist()
        self.df[selected_columns] = self.df[copied_columns]
        selected_columns = self.df.loc[:, 'WB_m0_w0_r0_f0':
                                       'WB_m18_w3_r3_f5'].columns.tolist()
        self.df[selected_columns] = self.df[copied_columns]
        selected_columns = self.df.loc[:, 'WPB_m0_w0_r0_f0':
                                       'WPB_m18_w3_r3_f5'].columns.tolist()
        self.df[selected_columns] = self.df[copied_columns]
        selected_columns = self.df.loc[:, 'EHP_m0_w0_r0_f0':
                                       'EHP_m18_w3_r3_f5'].columns.tolist()
        self.df[selected_columns] = self.df[copied_columns]
        return self.df
```
