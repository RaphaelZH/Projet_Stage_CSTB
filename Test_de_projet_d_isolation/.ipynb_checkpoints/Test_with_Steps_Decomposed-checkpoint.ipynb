class Heating_System:
    def __init__(self, df):
        self.df_heating_system = df

    def get_Heating_System_Type(self, t_switch, df):
        self.heating_system_to_switch = self.Heating_System(t_switch, df)[0]
        return self.heating_system_to_switch

    def get_Efficiency(self, t_switch, df):
        self.efficiency = float(self.Heating_System(t_switch, df)[1])
        return self.efficiency

    def get_A_Investment(self, t_switch, df):
        self.a_investment = float(self.Heating_System(t_switch, df)[2])
        return self.a_investment

    def get_B_Investment(self, t_switch, df):
        self.b_investment = float(self.Heating_System(t_switch, df)[3])
        return self.b_investment

    def get_Maintenance(self, t_switch, df):
        self.maintenance = float(self.Heating_System(t_switch, df)[4])
        return self.maintenance

        self.emission_factor = float(self.Heating_System(t_switch, df)[5])
        return self.emission_factor

class Collective_Dwelling_CI_Heating_System(Heating_System):
    def __init__(self, df):
        super().__init__(df)
        self.df_CD_CI_heating_system = self.df_heating_system.iloc[1:10, :]

    def Heating_System(self, t_switch, df):
        if t_switch == 'ER':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[0].tolist()
        elif t_switch == 'GB':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[1].tolist()
        elif t_switch == 'CGB':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[2].tolist()
        elif t_switch == 'OB':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[3].tolist()
        elif t_switch == 'COB':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[4].tolist()
        elif t_switch == 'EHP2':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[5].tolist()
        elif t_switch == 'WB':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[6].tolist()
        elif t_switch == 'WPB':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[7].tolist()
        elif t_switch == 'EHP':
            self.heating_syetem = self.df_CD_CI_heating_system.iloc[8].tolist()
        else:
            raise Exception('No matching heating system!')
        return self.heating_syetem

        self.surface_composant_w = self.share_of_windows * math.sqrt(

        return self.surface_composant_w

    def Ug(self, p, s, df):
        ug_1 = (
            (1 - self.share_of_windows) * self.um +
            self.share_of_windows * self.uw) * math.sqrt(
                self.surface / (self.floor + 1)) * 4 * self.ceiling_height * (
                    self.floor + 1) + (self.ur + self.uf) * self.surface / (
                        self.floor + 1)

        ug_2 = math.sqrt(
            self.surface / (self.floor + 1)) * 4 * self.ceiling_height * (
                self.floor + 1) + 2 * self.surface / (self.floor + 1)
        self.ug = ug_1 / ug_2
        return self.ug

    def UEN_per_Metre2(self, p, s, df):
        uen_per_metre2_1 = math.sqrt(
            self.surface / (self.floor + 1)) * 4 * self.ceiling_height * (
                self.floor + 1) + 2 * self.surface / (self.floor + 1)
        uen_per_metre2_2 = self.ug * self.intermittence * self.climate_zone / (
            self.surface * math.pow(10, 3))
        uen_per_metre2_3 = 1 + 0.1 * (
            (math.sqrt(self.surface /
                       (self.floor + 1)) * 4 * self.ceiling_height *
             (self.floor + 1) + 2 * self.surface /
             (self.floor + 1)) * self.ug /
            (self.surface * self.ceiling_height / 2.5) - 1)
        self.uen_per_metre2 = uen_per_metre2_1 * uen_per_metre2_2 / uen_per_metre2_3
        return self.uen_per_metre2
