## `Heating_System`

```python
class Heating_System:
    def __init__(self, df):
        self.df_heating_system = df

    def get_Heating_System_Type(self, t_switch, df):
        self.heating_system_to_switch = self.Heating_System(t_switch, df)[0]
        return self.heating_system_to_switch

    def get_Efficiency(self, t_switch, df):
        self.efficiency = float(self.Heating_System(t_switch, df)[1])
        return self.efficiency

    def get_A_Investment(self, t_switch, df):
        self.a_investment = float(self.Heating_System(t_switch, df)[2])
        return self.a_investment

    def get_B_Investment(self, t_switch, df):
        self.b_investment = float(self.Heating_System(t_switch, df)[3])
        return self.b_investment

    def get_Maintenance(self, t_switch, df):
        self.maintenance = float(self.Heating_System(t_switch, df)[4])
        return self.maintenance

    def get_Emission_Factor(self, t_switch, df):
        self.emission_factor = float(self.Heating_System(t_switch, df)[5])
        return self.emission_factor

    def get_Energy_Price_v1(self, t_switch, df):
        self.energy_price_v1 = float(self.Heating_System(t_switch, df)[6])
        return self.energy_price_v1

    def get_Same_Switch_Cost(self, t_switch, df):
        self.same_switch_cost = float(self.Heating_System(t_switch, df)[7])
        return self.same_switch_cost

    def get_Different_Switch_Cost(self, t_switch, df):
        self.different_switch_cost = float(
            self.Heating_System(t_switch, df)[8])
        return self.different_switch_cost

    def get_Delevery_Type_Cost(self, t_switch, df):
        self.delevery_type_cost = float(self.Heating_System(t_switch, df)[9])
        return self.delevery_type_cost

    def get_Specific_Gas_Price(self, t_switch, df):
        self.specific_gas_price = float(self.Heating_System(t_switch, df)[10])
        return self.specific_gas_price
```

## `Housing`

```python
class Housing:
    def __init__(self, df):
        self.df_housing = df
        self.carbon_price = 25
        self.discount_rate = 0.05
        self.annual_variation = 0

    def get_Housing_Type(self, p, s, df):
        self.housing_type = self.housing[0]
        return self.housing_type

    def get_Construction_Period(self, p, s, df):
        self.construction_period = self.housing[1]
        return self.construction_period

    def get_Size_Factor(self, p, s, df):
        self.size_factor = self.housing[2]
        return self.size_factor

    def get_Surface(self, p, s, df):
        self.surface = float(self.housing[3])
        return self.surface

    def get_Ceiling_Height(self, p, s, df):
        self.ceiling_height = float(self.housing[4])
        return self.ceiling_height

    def get_Share_of_Windows(self, p, s, df):
        self.share_of_windows = float(self.housing[5])
        return self.share_of_windows

    def get_Dwelling_per_Floor(self, p, s, df):
        self.dwelling_per_floor = int(self.housing[6])
        return self.dwelling_per_floor

    def get_Floor(self, p, s, df):
        self.floor = int(self.housing[7])
        return self.floor

    def get_Intermittence(self, p, s, df):
        self.intermittence = float(self.housing[8])
        return self.intermittence

    def get_Um(self, p, s, df):
        self.um = float(self.housing[9])
        return self.um

    def get_Uw(self, p, s, df):
        self.uw = float(self.housing[10])
        return self.uw

    def get_Ur(self, p, s, df):
        self.ur = float(self.housing[11])
        return self.ur

    def get_Uf(self, p, s, df):
        self.uf = float(self.housing[12])
        return self.uf

    def get_Climate_Zone(self, p, s, df):
        self.climate_zone = int(self.housing[13])
        return self.climate_zone

    def Surface_Composant_r(self, p, s, df):
        self.surface_composant_r = self.surface / (self.floor + 1)
        return self.surface_composant_r

    def Surface_Composant_f(self, p, s, df):
        self.surface_composant_f = self.surface / (self.floor + 1)
        return self.surface_composant_f

    def Surface_Composant_total(self, p, s, df):
        self.surface_composant_total = self.surface_composant_m + self.surface_composant_w + self.surface_composant_r + self.surface_composant_f
        return self.surface_composant_total

    def UEN_per_Dwelling(self, p, s, df):
        self.uen_per_dwelling = self.uen_per_metre2 * self.surface
        return self.uen_per_dwelling
```

### `Collective_Dwelling_CI(Housing)`

```python
class Collective_Dwelling_CI(Housing):
    def __init__(self, df):
        super().__init__(df)
        self.df_Collective_Dwelling_CI = self.df_housing.iloc[22:43, :]

    def Housing_Categorizer(self, p, s, df):
        if p == "<1948":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CI.iloc[0].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CI.iloc[1].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CI.iloc[2].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1948-1974":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CI.iloc[3].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CI.iloc[4].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CI.iloc[5].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1975-1981":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CI.iloc[6].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CI.iloc[7].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CI.iloc[8].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1982-1989":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CI.iloc[9].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CI.iloc[10].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CI.iloc[11].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1990-2000":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CI.iloc[12].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CI.iloc[13].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CI.iloc[14].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "2001-2005":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CI.iloc[15].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CI.iloc[16].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CI.iloc[17].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "2006-2013":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CI.iloc[18].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CI.iloc[19].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CI.iloc[20].tolist()
            else:
                raise Exception('No matching size factor!')
        else:
            raise Exception('No matching construction period!')
        return self.housing

    def Surface_Composant_m(self, p, s, df):
        self.surface_composant_m_building = (
            1 - self.share_of_windows) * math.sqrt(
                self.surface * self.dwelling_per_floor) * 4 * (
                    self.floor + 1) * self.ceiling_height
        self.surface_composant_m = self.surface_composant_m_building / (
            (self.floor + 1) * self.dwelling_per_floor)
        return self.surface_composant_m

    def Surface_Composant_w(self, p, s, df):
        self.surface_composant_w_building = self.share_of_windows * math.sqrt(
            self.surface * self.dwelling_per_floor) * 4 * (
                self.floor + 1) * self.ceiling_height
        self.surface_composant_w = self.surface_composant_w_building / (
            (self.floor + 1) * self.dwelling_per_floor)
        return self.surface_composant_w

    def Ug(self, p, s, df):
        ug_1 = (1 - self.share_of_windows) * self.um * math.sqrt(
            self.dwelling_per_floor * self.surface
        ) * 4 * (
            self.floor + 1
        ) * self.ceiling_height + self.share_of_windows * self.uw * math.sqrt(
            self.dwelling_per_floor * self.surface) * 4 * (
                self.floor + 1) * self.ceiling_height + (
                    self.uf + self.ur) * self.dwelling_per_floor * self.surface
        ug_2 = math.sqrt(self.dwelling_per_floor * self.surface) * 4 * (
            self.floor + 1
        ) * self.ceiling_height + 2 * self.dwelling_per_floor * self.surface
        self.ug = ug_1 / ug_2
        return self.ug

    def UEN_per_Metre2(self, p, s, df):
        uen_per_metre2_1 = self.ug * (
            self.dwelling_per_floor * self.surface * 2 +
            math.sqrt(self.dwelling_per_floor * self.surface) * 4 *
            (self.floor + 1) * self.ceiling_height
        ) * self.climate_zone * self.intermittence
        uen_per_metre2_2 = 1 + 0.1 * (
            self.ug * (self.dwelling_per_floor * self.surface * 2 +
                       math.sqrt(self.dwelling_per_floor * self.surface) * 4 *
                       (self.floor + 1) * self.ceiling_height) /
            (self.dwelling_per_floor * self.surface *
             (self.floor + 1) * self.ceiling_height / 2.5) - 1)
        uen_per_metre2_3 = self.dwelling_per_floor * self.surface * (
            self.floor + 1) * 1000
        self.uen_per_metre2 = uen_per_metre2_1 / uen_per_metre2_2 / uen_per_metre2_3
        return self.uen_per_metre2

    def Call_Functions(self, p, s, t_switch, df):
        fonctions_list = [
            self.Housing_Categorizer(p, s, df),
            self.get_Housing_Type(p, s, df),
            self.get_Construction_Period(p, s, df),
            self.get_Size_Factor(p, s, df),
            self.get_Surface(p, s, df),
            self.get_Ceiling_Height(p, s, df),
            self.get_Share_of_Windows(p, s, df),
            self.get_Dwelling_per_Floor(p, s, df),
            self.get_Floor(p, s, df),
            self.get_Intermittence(p, s, df),
            self.get_Um(p, s, df),
            self.get_Uw(p, s, df),
            self.get_Ur(p, s, df),
            self.get_Uf(p, s, df),
            self.get_Climate_Zone(p, s, df),
            self.Surface_Composant_m(p, s, df),
            self.Surface_Composant_w(p, s, df),
            self.Surface_Composant_f(p, s, df),
            self.Surface_Composant_r(p, s, df),
            self.Surface_Composant_total(p, s, df),
            self.Ug(p, s, df),
            self.UEN_per_Metre2(p, s, df),
            self.UEN_per_Dwelling(p, s, df),
        ]
        return fonctions_list

    def Initial_GHC(self, p, s, t_switch, df):
        self.Call_Functions(p, s, t_switch, df)
        self.df_CD_CI_heating_system = Collective_Dwelling_CI_Heating_System(
            df_Heating_System)
        self.heating_system_to_switch = self.df_CD_CI_heating_system.get_Heating_System_Type(
            t_switch, df_Heating_System)
        self.efficiency = self.df_CD_CI_heating_system.get_Efficiency(
            t_switch, df_Heating_System)
        self.a_investment = self.df_CD_CI_heating_system.get_A_Investment(
            t_switch, df_Heating_System)
        self.b_investment = self.df_CD_CI_heating_system.get_B_Investment(
            t_switch, df_Heating_System)
        self.maintenance = self.df_CD_CI_heating_system.get_Maintenance(
            t_switch, df_Heating_System)
        self.emission_factor = self.df_CD_CI_heating_system.get_Emission_Factor(
            t_switch, df_Heating_System)
        self.energy_price_v1 = self.df_CD_CI_heating_system.get_Energy_Price_v1(
            t_switch, df_Heating_System)

        initial_ghc_1 = (self.a_investment +
                         self.b_investment * self.uen_per_dwelling) / math.pow(
                             (1 + self.discount_rate), (19 - 0))
        initial_ghc_2 = (self.a_investment +
                         self.b_investment * self.uen_per_dwelling) / math.pow(
                             (1 + self.discount_rate), (39 - 0))
        initial_ghc_3 = self.uen_per_dwelling / self.efficiency
        initial_ghc_4 = 0
        self.price_1 = self.energy_price_v1 * (
            1 +
            self.annual_variation) + self.emission_factor * self.carbon_price
        for i in range(40):
            self.price = self.price_1 * math.pow(
                (1 + self.annual_variation), 1)
            initial_ghc_4 += self.price / math.pow(
                (1 + self.discount_rate), i + 1)
        initial_ghc_5 = self.maintenance
        self.insulation_cost = 0
        self.initial_ghc = initial_ghc_1 + initial_ghc_2 + \
        initial_ghc_3 * initial_ghc_4 + initial_ghc_5 + \
        self.insulation_cost
        return self.initial_ghc
```

#### `Collective_Dwelling_CI_Switchable_Insulation(Collective_Dwelling_CI)`

```python
class Collective_Dwelling_CI_Switchable_Insulation(Collective_Dwelling_CI):
    def __init__(self, df):
        super().__init__(df)

    def get_Um(self, p, s, m, df):
        if m == 0:
            self.um = float(self.Housing_Categorizer(p, s, df)[9])
        else:
            M = m + 19
            self.um = float(df_Wall_Investment.iloc[:, 2].tolist()[M])
        return self.um

    def get_Uw(self, p, s, w, df):
        if w == 0:
            self.uw = float(self.Housing_Categorizer(p, s, df)[10])
        else:
            W = w + 3
            self.uw = float(df_Windows_Investment.iloc[:, 2].tolist()[W])
        return self.uw

    def get_Ur(self, p, s, r, df):
        if r == 0:
            self.ur = float(self.Housing_Categorizer(p, s, df)[11])
        else:
            R = r + 7
            self.ur = float(df_Roof_Investment.iloc[:, 2].tolist()[R])
        return self.ur

    def get_Uf(self, p, s, f, df):
        if f == 0:
            self.uf = float(self.Housing_Categorizer(p, s, df)[12])
        else:
            F = f + 3
            self.uf = float(df_Floor_Investment.iloc[:, 2].tolist()[F])
        return self.uf

    def Ug(self, p, s, m, w, r, f, df):
        ug_1 = (1 - self.share_of_windows) * self.um * math.sqrt(
            self.dwelling_per_floor * self.surface
        ) * 4 * (
            self.floor + 1
        ) * self.ceiling_height + self.share_of_windows * self.uw * math.sqrt(
            self.dwelling_per_floor * self.surface) * 4 * (
                self.floor + 1) * self.ceiling_height + (
                    self.uf + self.ur) * self.dwelling_per_floor * self.surface
        ug_2 = math.sqrt(self.dwelling_per_floor * self.surface) * 4 * (
            self.floor + 1
        ) * self.ceiling_height + 2 * self.dwelling_per_floor * self.surface
        self.ug = ug_1 / ug_2
        return self.ug

    def UEN_per_Metre2(self, p, s, m, w, r, f, df):
        uen_per_metre2_1 = self.ug * (
            self.dwelling_per_floor * self.surface * 2 +
            math.sqrt(self.dwelling_per_floor * self.surface) * 4 *
            (self.floor + 1) * self.ceiling_height
        ) * self.climate_zone * self.intermittence
        uen_per_metre2_2 = 1 + 0.1 * (
            self.ug * (self.dwelling_per_floor * self.surface * 2 +
                       math.sqrt(self.dwelling_per_floor * self.surface) * 4 *
                       (self.floor + 1) * self.ceiling_height) /
            (self.dwelling_per_floor * self.surface *
             (self.floor + 1) * self.ceiling_height / 2.5) - 1)
        uen_per_metre2_3 = self.dwelling_per_floor * self.surface * (
            self.floor + 1) * 1000
        self.uen_per_metre2 = uen_per_metre2_1 / uen_per_metre2_2 / uen_per_metre2_3
        return self.uen_per_metre2

    def UEN_per_Dwelling(self, p, s, m, w, r, f, df):
        self.uen_per_dwelling = self.uen_per_metre2 * self.surface
        return self.uen_per_dwelling

    def Call_Functions_Switchable(self, p, s, m, w, r, f, df):
        fonctions_list = [
            self.Housing_Categorizer(p, s, df),
            self.get_Housing_Type(p, s, df),
            self.get_Construction_Period(p, s, df),
            self.get_Size_Factor(p, s, df),
            self.get_Surface(p, s, df),
            self.get_Ceiling_Height(p, s, df),
            self.get_Share_of_Windows(p, s, df),
            self.get_Dwelling_per_Floor(p, s, df),
            self.get_Floor(p, s, df),
            self.get_Intermittence(p, s, df),
            self.get_Um(p, s, m, df),
            self.get_Uw(p, s, w, df),
            self.get_Ur(p, s, r, df),
            self.get_Uf(p, s, f, df),
            self.get_Climate_Zone(p, s, df),
            self.Surface_Composant_m(p, s, df),
            self.Surface_Composant_w(p, s, df),
            self.Surface_Composant_f(p, s, df),
            self.Surface_Composant_r(p, s, df),
            self.Surface_Composant_total(p, s, df),
            self.Ug(p, s, m, w, r, f, df),
            self.UEN_per_Metre2(p, s, m, w, r, f, df),
            self.UEN_per_Dwelling(p, s, m, w, r, f, df)
        ]
        return fonctions_list

    def GHC_Swithable(self, p, s, t_initial, t_switch, m, w, r, f, df):
        self.Call_Functions_Switchable(p, s, m, w, r, f, df)

        self.df_CD_CI_heating_system = Collective_Dwelling_CI_Heating_System(
            df_Heating_System)

        self.heating_system_to_switch = self.df_CD_CI_heating_system.get_Heating_System_Type(
            t_switch, df_Heating_System)

        self.initial_heating_system = self.df_CD_CI_heating_system.get_Heating_System_Type(
            t_initial, df_Heating_System)

        self.efficiency = self.df_CD_CI_heating_system.get_Efficiency(
            t_switch, df_Heating_System)
        self.a_investment = self.df_CD_CI_heating_system.get_A_Investment(
            t_switch, df_Heating_System)
        self.b_investment = self.df_CD_CI_heating_system.get_B_Investment(
            t_switch, df_Heating_System)
        self.maintenance = self.df_CD_CI_heating_system.get_Maintenance(
            t_switch, df_Heating_System)
        self.emission_factor = self.df_CD_CI_heating_system.get_Emission_Factor(
            t_switch, df_Heating_System)
        self.energy_price_v1 = self.df_CD_CI_heating_system.get_Energy_Price_v1(
            t_switch, df_Heating_System)

        self.same_switch_cost = self.df_CD_CI_heating_system.get_Same_Switch_Cost(
            t_switch, df_Heating_System)
        self.different_switch_cost = self.df_CD_CI_heating_system.get_Different_Switch_Cost(
            t_switch, df_Heating_System)
        self.delevery_type_cost = self.df_CD_CI_heating_system.get_Delevery_Type_Cost(
            t_switch, df_Heating_System)
        self.specific_gas_price = self.df_CD_CI_heating_system.get_Specific_Gas_Price(
            t_switch, df_Heating_System)

        CD_CI_initial_ghc_1 = (self.a_investment + self.b_investment *
                               self.uen_per_dwelling) / math.pow(
                                   (1 + self.discount_rate), 0)
        CD_CI_initial_ghc_2 = (self.a_investment + self.b_investment *
                               self.uen_per_dwelling) / math.pow(
                                   (1 + self.discount_rate), 20)
        CD_CI_initial_ghc_3 = self.uen_per_dwelling / self.efficiency
        CD_CI_initial_ghc_4 = 0
        CD_CI_initial_ghc_5 = self.maintenance

        if m == 0:
            self.CD_CI_cost_addition_m = self.surface_composant_m * 0
        else:
            M = m + 19
            self.CD_CI_cost_addition_m = self.surface_composant_m * float(
                df_Wall_Investment.iloc[:, 3].tolist()[M])

        if w == 0:
            self.CD_CI_cost_addition_w = self.surface_composant_w * 0
        else:
            W = w + 3
            self.CD_CI_cost_addition_w = self.surface_composant_w * float(
                df_Windows_Investment.iloc[:, 3].tolist()[W])

        if r == 0:
            self.CD_CI_cost_addition_r = self.surface_composant_r * 0
        else:
            R = r + 7
            self.CD_CI_cost_addition_r = self.surface_composant_r * float(
                df_Roof_Investment.iloc[:, 3].tolist()[R])

        if f == 0:
            self.CD_CI_cost_addition_f = self.surface_composant_f * 0
        else:
            F = f + 3
            self.CD_CI_cost_addition_f = self.surface_composant_f * float(
                df_Floor_Investment.iloc[:, 3].tolist()[F])

        self.CD_CI_insulation_cost = self.CD_CI_cost_addition_m + self.CD_CI_cost_addition_w + \
        self.CD_CI_cost_addition_r + self.CD_CI_cost_addition_f

        self.centralized_cost = self.delevery_type_cost * self.surface
        self.decentralized_cost = self.delevery_type_cost * self.surface
        self.control_cost = 0

        self.price_1 = self.energy_price_v1 * (
            1 +
            self.annual_variation) + self.emission_factor * self.carbon_price

        if self.heating_system_to_switch == self.initial_heating_system:

            self.specific_cost = 0

            CD_CI_initial_ghc_1 = (self.a_investment + self.b_investment *
                                   self.uen_per_dwelling) / math.pow(
                                       (1 + self.discount_rate), (19 - 0))
            CD_CI_initial_ghc_2 = (self.a_investment + self.b_investment *
                                   self.uen_per_dwelling) / math.pow(
                                       (1 + self.discount_rate), (39 - 0))
        else:
            if self.heating_system_to_switch in [
                    'CI Electric Heater (ER)', 'CI Wood Stove (WB)',
                    'CI Air Source Electric Heat Pump (EHP)'
            ]:
                if self.initial_heating_system in ['CI Gas Boiler (GB)', 'CI Condensing Gas Boiler (CGB)', 'CI Oil Boiler (OB)', \
                                           'CI Condensing Oil Boiler (COB)', 'CI Water Source Electric Heat Pump (EHP2)', \
                                           'CI Wood Pellet Boiler (WPB)']:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.decentralized_cost

                else:

                    self.specific_cost = self.same_switch_cost

            elif self.heating_system_to_switch in [
                    'CI Gas Boiler (GB)', 'CI Condensing Gas Boiler (CGB)'
            ]:

                if self.initial_heating_system in [
                        'CI Gas Boiler (GB)', 'CI Condensing Gas Boiler (CGB)'
                ]:

                    self.specific_cost = self.same_switch_cost

                elif self.initial_heating_system in [
                        'CI Electric Heater (ER)', 'CI Wood Stove (WB)',
                        'CI Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                elif self.initial_heating_system in [
                        'CI Oil Boiler (OB)', 'CI Condensing Oil Boiler (COB)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.price_1 = self.specific_gas_price * (
                        1 + self.annual_variation
                    ) + self.emission_factor * self.carbon_price

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch in [
                    'CI Oil Boiler (OB)', 'CI Condensing Oil Boiler (COB)'
            ]:

                if self.initial_heating_system in [
                        'CI Oil Boiler (OB)', 'CI Condensing Oil Boiler (COB)'
                ]:

                    self.specific_cost = self.same_switch_cost

                elif self.initial_heating_system in [
                        'CI Electric Heater (ER)', 'CI Wood Stove (WB)',
                        'CI Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch == 'CI Wood Pellet Boiler (WPB)':

                if self.initial_heating_system == 'CI Wood Stove (WB)':

                    self.specific_cost = self.same_switch_cost

                    self.control_cost = self.centralized_cost

                elif self.initial_heating_system in [
                        'CI Electric Heater (ER)',
                        'CI Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

            else:

                if self.initial_heating_system in [
                        'CI Electric Heater (ER)', 'CI Wood Stove (WB)',
                        'CI Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

        for i in range(40):
            self.price = self.price_1 * math.pow(
                (1 + self.annual_variation), 1)
            CD_CI_initial_ghc_4 += self.price / math.pow(
                (1 + self.discount_rate), i + 1)

        self.CD_CI_ghc_swithable = CD_CI_initial_ghc_1 + CD_CI_initial_ghc_2 + \
        CD_CI_initial_ghc_3 * CD_CI_initial_ghc_4 + \
        CD_CI_initial_ghc_5 + self.CD_CI_insulation_cost + self.specific_cost + \
        self.control_cost
        return self.CD_CI_ghc_swithable
```

```python
class Collective_Dwelling_CI_per_Year_of_HS(
        Collective_Dwelling_CI_Switchable_Insulation):
    def __init__(self, df):
        super().__init__(df)

    def Call_Functions_Switchable(self, p, s, m, w, r, f, df):
        fonctions_list = [
            self.Housing_Categorizer(p, s, df),
            self.get_Housing_Type(p, s, df),
            self.get_Construction_Period(p, s, df),
            self.get_Size_Factor(p, s, df),
            self.get_Surface(p, s, df),
            self.get_Ceiling_Height(p, s, df),
            self.get_Share_of_Windows(p, s, df),
            self.get_Dwelling_per_Floor(p, s, df),
            self.get_Floor(p, s, df),
            self.get_Intermittence(p, s, df),
            self.get_Um(p, s, m, df),
            self.get_Uw(p, s, w, df),
            self.get_Ur(p, s, r, df),
            self.get_Uf(p, s, f, df),
            self.get_Climate_Zone(p, s, df),
            self.Surface_Composant_m(p, s, df),
            self.Surface_Composant_w(p, s, df),
            self.Surface_Composant_f(p, s, df),
            self.Surface_Composant_r(p, s, df),
            self.Surface_Composant_total(p, s, df),
            self.Ug(p, s, m, w, r, f, df),
            self.UEN_per_Metre2(p, s, m, w, r, f, df),
            self.UEN_per_Dwelling(p, s, m, w, r, f, df)
        ]
        return fonctions_list

    def GHC_Swithable_per_Year(self, p, s, t_initial, t_switch, m, w, r, f, df,
                               year):

        self.Call_Functions_Switchable(p, s, m, w, r, f, df)

        self.df_CD_CI_heating_system = Collective_Dwelling_CI_Heating_System(
            df_Heating_System)

        self.heating_system_to_switch = self.df_CD_CI_heating_system.get_Heating_System_Type(
            t_switch, df_Heating_System)

        self.initial_heating_system = self.df_CD_CI_heating_system.get_Heating_System_Type(
            t_initial, df_Heating_System)

        self.efficiency = self.df_CD_CI_heating_system.get_Efficiency(
            t_switch, df_Heating_System)
        self.a_investment = self.df_CD_CI_heating_system.get_A_Investment(
            t_switch, df_Heating_System)
        self.b_investment = self.df_CD_CI_heating_system.get_B_Investment(
            t_switch, df_Heating_System)
        self.maintenance = self.df_CD_CI_heating_system.get_Maintenance(
            t_switch, df_Heating_System)
        self.emission_factor = self.df_CD_CI_heating_system.get_Emission_Factor(
            t_switch, df_Heating_System)
        self.energy_price_v1 = self.df_CD_CI_heating_system.get_Energy_Price_v1(
            t_switch, df_Heating_System)

        self.same_switch_cost = self.df_CD_CI_heating_system.get_Same_Switch_Cost(
            t_switch, df_Heating_System)
        self.different_switch_cost = self.df_CD_CI_heating_system.get_Different_Switch_Cost(
            t_switch, df_Heating_System)
        self.delevery_type_cost = self.df_CD_CI_heating_system.get_Delevery_Type_Cost(
            t_switch, df_Heating_System)
        self.specific_gas_price = self.df_CD_CI_heating_system.get_Specific_Gas_Price(
            t_switch, df_Heating_System)

        CD_CI_initial_ghc_1 = (self.a_investment + self.b_investment *
                               self.uen_per_dwelling) / math.pow(
                                   (1 + self.discount_rate), 0)
        CD_CI_initial_ghc_2 = (self.a_investment + self.b_investment *
                               self.uen_per_dwelling) / math.pow(
                                   (1 + self.discount_rate), 20)
        CD_CI_initial_ghc_3 = self.uen_per_dwelling / self.efficiency
        CD_CI_initial_ghc_4 = 0
        CD_CI_initial_ghc_5 = self.maintenance

        if m == 0:
            self.CD_CI_cost_addition_m = self.surface_composant_m * 0
        else:
            M = m + 19
            self.CD_CI_cost_addition_m = self.surface_composant_m * float(
                df_Wall_Investment.iloc[:, 3].tolist()[M])

        if w == 0:
            self.CD_CI_cost_addition_w = self.surface_composant_w * 0
        else:
            W = w + 3
            self.CD_CI_cost_addition_w = self.surface_composant_w * float(
                df_Windows_Investment.iloc[:, 3].tolist()[W])

        if r == 0:
            self.CD_CI_cost_addition_r = self.surface_composant_r * 0
        else:
            R = r + 7
            self.CD_CI_cost_addition_r = self.surface_composant_r * float(
                df_Roof_Investment.iloc[:, 3].tolist()[R])

        if f == 0:
            self.CD_CI_cost_addition_f = self.surface_composant_f * 0
        else:
            F = f + 3
            self.CD_CI_cost_addition_f = self.surface_composant_f * float(
                df_Floor_Investment.iloc[:, 3].tolist()[F])

        self.CD_CI_insulation_cost = self.CD_CI_cost_addition_m + self.CD_CI_cost_addition_w + \
        self.CD_CI_cost_addition_r + self.CD_CI_cost_addition_f

        self.centralized_cost = self.delevery_type_cost * self.surface
        self.decentralized_cost = self.delevery_type_cost * self.surface
        self.control_cost = 0

        self.price_1 = self.energy_price_v1 * (
            1 +
            self.annual_variation) + self.emission_factor * self.carbon_price

        if self.heating_system_to_switch == self.initial_heating_system:

            self.specific_cost = 0

            CD_CI_initial_ghc_1 = (self.a_investment + self.b_investment *
                                   self.uen_per_dwelling) / math.pow(
                                       (1 + self.discount_rate), (19 - year))
            CD_CI_initial_ghc_2 = (self.a_investment + self.b_investment *
                                   self.uen_per_dwelling) / math.pow(
                                       (1 + self.discount_rate), (39 - year))
        else:
            if self.heating_system_to_switch in [
                    'CI Electric Heater (ER)', 'CI Wood Stove (WB)',
                    'CI Air Source Electric Heat Pump (EHP)'
            ]:
                if self.initial_heating_system in ['CI Gas Boiler (GB)', 'CI Condensing Gas Boiler (CGB)', 'CI Oil Boiler (OB)', \
                                           'CI Condensing Oil Boiler (COB)', 'CI Water Source Electric Heat Pump (EHP2)', \
                                           'CI Wood Pellet Boiler (WPB)']:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.decentralized_cost

                else:

                    self.specific_cost = self.same_switch_cost

            elif self.heating_system_to_switch in [
                    'CI Gas Boiler (GB)', 'CI Condensing Gas Boiler (CGB)'
            ]:

                if self.initial_heating_system in [
                        'CI Gas Boiler (GB)', 'CI Condensing Gas Boiler (CGB)'
                ]:

                    self.specific_cost = self.same_switch_cost

                elif self.initial_heating_system in [
                        'CI Electric Heater (ER)', 'CI Wood Stove (WB)',
                        'CI Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                elif self.initial_heating_system in [
                        'CI Oil Boiler (OB)', 'CI Condensing Oil Boiler (COB)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.price_1 = self.specific_gas_price * (
                        1 + self.annual_variation
                    ) + self.emission_factor * self.carbon_price

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch in [
                    'CI Oil Boiler (OB)', 'CI Condensing Oil Boiler (COB)'
            ]:

                if self.initial_heating_system in [
                        'CI Oil Boiler (OB)', 'CI Condensing Oil Boiler (COB)'
                ]:

                    self.specific_cost = self.same_switch_cost

                elif self.initial_heating_system in [
                        'CI Electric Heater (ER)', 'CI Wood Stove (WB)',
                        'CI Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch == 'CI Wood Pellet Boiler (WPB)':

                if self.initial_heating_system == 'CI Wood Stove (WB)':

                    self.specific_cost = self.same_switch_cost

                    self.control_cost = self.centralized_cost

                elif self.initial_heating_system in [
                        'CI Electric Heater (ER)',
                        'CI Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

            else:

                if self.initial_heating_system in [
                        'CI Electric Heater (ER)', 'CI Wood Stove (WB)',
                        'CI Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

        for i in range(40):
            self.price = self.price_1 * math.pow(
                (1 + self.annual_variation), 1)
            CD_CI_initial_ghc_4 += self.price / math.pow(
                (1 + self.discount_rate), i + 1)

        self.CD_CI_ghc_swithable = CD_CI_initial_ghc_1 + CD_CI_initial_ghc_2 + \
        CD_CI_initial_ghc_3 * CD_CI_initial_ghc_4 + \
        CD_CI_initial_ghc_5 + self.CD_CI_insulation_cost + self.specific_cost + \
        self.control_cost
        return self.CD_CI_ghc_swithable
```

### `Collective_Dwelling_CC(Housing)`

```python
class Collective_Dwelling_CC(Housing):
    def __init__(self, df):
        super().__init__(df)
        self.df_Collective_Dwelling_CC = self.df_housing.iloc[22:43, :]

    def Housing_Categorizer(self, p, s, df):
        if p == "<1948":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CC.iloc[0].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CC.iloc[1].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CC.iloc[2].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1948-1974":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CC.iloc[3].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CC.iloc[4].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CC.iloc[5].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1975-1981":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CC.iloc[6].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CC.iloc[7].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CC.iloc[8].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1982-1989":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CC.iloc[9].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CC.iloc[10].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CC.iloc[11].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1990-2000":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CC.iloc[12].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CC.iloc[13].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CC.iloc[14].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "2001-2005":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CC.iloc[15].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CC.iloc[16].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CC.iloc[17].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "2006-2013":
            if s == "Middle":
                self.housing = self.df_Collective_Dwelling_CC.iloc[18].tolist()
            elif s == "Small":
                self.housing = self.df_Collective_Dwelling_CC.iloc[19].tolist()
            elif s == "Large":
                self.housing = self.df_Collective_Dwelling_CC.iloc[20].tolist()
            else:
                raise Exception('No matching size factor!')
        else:
            raise Exception('No matching construction period!')
        return self.housing

    def Surface_Composant_m(self, p, s, df):
        self.surface_composant_m_building = (
            1 - self.share_of_windows) * math.sqrt(
                self.surface * self.dwelling_per_floor) * 4 * (
                    self.floor + 1) * self.ceiling_height
        self.surface_composant_m = self.surface_composant_m_building / (
            (self.floor + 1) * self.dwelling_per_floor)
        return self.surface_composant_m

    def Surface_Composant_w(self, p, s, df):
        self.surface_composant_w_building = self.share_of_windows * math.sqrt(
            self.surface * self.dwelling_per_floor) * 4 * (
                self.floor + 1) * self.ceiling_height
        self.surface_composant_w = self.surface_composant_w_building / (
            (self.floor + 1) * self.dwelling_per_floor)
        return self.surface_composant_w

    def Ug(self, p, s, df):
        ug_1 = (1 - self.share_of_windows) * self.um * math.sqrt(
            self.dwelling_per_floor * self.surface
        ) * 4 * (
            self.floor + 1
        ) * self.ceiling_height + self.share_of_windows * self.uw * math.sqrt(
            self.dwelling_per_floor * self.surface) * 4 * (
                self.floor + 1) * self.ceiling_height + (
                    self.uf + self.ur) * self.dwelling_per_floor * self.surface
        ug_2 = math.sqrt(self.dwelling_per_floor * self.surface) * 4 * (
            self.floor + 1
        ) * self.ceiling_height + 2 * self.dwelling_per_floor * self.surface
        self.ug = ug_1 / ug_2
        return self.ug

    def UEN_per_Metre2(self, p, s, df):
        uen_per_metre2_1 = self.ug * (
            self.dwelling_per_floor * self.surface * 2 +
            math.sqrt(self.dwelling_per_floor * self.surface) * 4 *
            (self.floor + 1) * self.ceiling_height
        ) * self.climate_zone * self.intermittence
        uen_per_metre2_2 = 1 + 0.1 * (
            self.ug * (self.dwelling_per_floor * self.surface * 2 +
                       math.sqrt(self.dwelling_per_floor * self.surface) * 4 *
                       (self.floor + 1) * self.ceiling_height) /
            (self.dwelling_per_floor * self.surface *
             (self.floor + 1) * self.ceiling_height / 2.5) - 1)
        uen_per_metre2_3 = self.dwelling_per_floor * self.surface * (
            self.floor + 1) * 1000
        self.uen_per_metre2 = uen_per_metre2_1 / uen_per_metre2_2 / uen_per_metre2_3
        return self.uen_per_metre2

    def Call_Functions(self, p, s, t_switch, df):
        fonctions_list = [
            self.Housing_Categorizer(p, s, df),
            self.get_Housing_Type(p, s, df),
            self.get_Construction_Period(p, s, df),
            self.get_Size_Factor(p, s, df),
            self.get_Surface(p, s, df),
            self.get_Ceiling_Height(p, s, df),
            self.get_Share_of_Windows(p, s, df),
            self.get_Dwelling_per_Floor(p, s, df),
            self.get_Floor(p, s, df),
            self.get_Intermittence(p, s, df),
            self.get_Um(p, s, df),
            self.get_Uw(p, s, df),
            self.get_Ur(p, s, df),
            self.get_Uf(p, s, df),
            self.get_Climate_Zone(p, s, df),
            self.Surface_Composant_m(p, s, df),
            self.Surface_Composant_w(p, s, df),
            self.Surface_Composant_f(p, s, df),
            self.Surface_Composant_r(p, s, df),
            self.Surface_Composant_total(p, s, df),
            self.Ug(p, s, df),
            self.UEN_per_Metre2(p, s, df),
            self.UEN_per_Dwelling(p, s, df),
        ]
        return fonctions_list

    def Initial_GHC(self, p, s, t_switch, df):
        self.Call_Functions(p, s, t_switch, df)
        self.df_CD_CC_heating_system = Collective_Dwelling_CC_Heating_System(
            df_Heating_System)
        self.heating_system_to_switch = self.df_CD_CC_heating_system.get_Heating_System_Type(
            t_switch, df_Heating_System)
        self.efficiency = self.df_CD_CC_heating_system.get_Efficiency(
            t_switch, df_Heating_System)
        self.a_investment = self.df_CD_CC_heating_system.get_A_Investment(
            t_switch, df_Heating_System)
        self.b_investment = self.df_CD_CC_heating_system.get_B_Investment(
            t_switch, df_Heating_System)
        self.maintenance = self.df_CD_CC_heating_system.get_Maintenance(
            t_switch, df_Heating_System)
        self.emission_factor = self.df_CD_CC_heating_system.get_Emission_Factor(
            t_switch, df_Heating_System)
        self.energy_price_v1 = self.df_CD_CC_heating_system.get_Energy_Price_v1(
            t_switch, df_Heating_System)

        initial_ghc_1 = (self.a_investment +
                         self.b_investment * self.uen_per_dwelling) / math.pow(
                             (1 + self.discount_rate), (19 - 0))
        initial_ghc_2 = (self.a_investment +
                         self.b_investment * self.uen_per_dwelling) / math.pow(
                             (1 + self.discount_rate), (39 - 0))
        initial_ghc_3 = self.uen_per_dwelling / self.efficiency
        initial_ghc_4 = 0
        self.price_1 = self.energy_price_v1 * (
            1 +
            self.annual_variation) + self.emission_factor * self.carbon_price
        for i in range(40):
            self.price = self.price_1 * math.pow(
                (1 + self.annual_variation), 1)
            initial_ghc_4 += self.price / math.pow(
                (1 + self.discount_rate), i + 1)
        initial_ghc_5 = self.maintenance
        self.insulation_cost = 0
        self.initial_ghc = initial_ghc_1 + initial_ghc_2 + \
        initial_ghc_3 * initial_ghc_4 + initial_ghc_5 + \
        self.insulation_cost
        return self.initial_ghc
```

```python
class Collective_Dwelling_CC_Switchable_Insulation(Collective_Dwelling_CC):
    def __init__(self, df):
        super().__init__(df)

    def get_Um(self, p, s, m, df):
        if m == 0:
            self.um = float(self.Housing_Categorizer(p, s, df)[9])
        else:
            M = m + 19
            self.um = float(df_Wall_Investment.iloc[:, 2].tolist()[M])
        return self.um

    def get_Uw(self, p, s, w, df):
        if w == 0:
            self.uw = float(self.Housing_Categorizer(p, s, df)[10])
        else:
            W = w + 3
            self.uw = float(df_Windows_Investment.iloc[:, 2].tolist()[W])
        return self.uw

    def get_Ur(self, p, s, r, df):
        if r == 0:
            self.ur = float(self.Housing_Categorizer(p, s, df)[11])
        else:
            R = r + 7
            self.ur = float(df_Roof_Investment.iloc[:, 2].tolist()[R])
        return self.ur

    def get_Uf(self, p, s, f, df):
        if f == 0:
            self.uf = float(self.Housing_Categorizer(p, s, df)[12])
        else:
            F = f + 3
            self.uf = float(df_Floor_Investment.iloc[:, 2].tolist()[F])
        return self.uf

    def Ug(self, p, s, m, w, r, f, df):
        ug_1 = (1 - self.share_of_windows) * self.um * math.sqrt(
            self.dwelling_per_floor * self.surface
        ) * 4 * (
            self.floor + 1
        ) * self.ceiling_height + self.share_of_windows * self.uw * math.sqrt(
            self.dwelling_per_floor * self.surface) * 4 * (
                self.floor + 1) * self.ceiling_height + (
                    self.uf + self.ur) * self.dwelling_per_floor * self.surface
        ug_2 = math.sqrt(self.dwelling_per_floor * self.surface) * 4 * (
            self.floor + 1
        ) * self.ceiling_height + 2 * self.dwelling_per_floor * self.surface
        self.ug = ug_1 / ug_2
        return self.ug

    def UEN_per_Metre2(self, p, s, m, w, r, f, df):
        uen_per_metre2_1 = self.ug * (
            self.dwelling_per_floor * self.surface * 2 +
            math.sqrt(self.dwelling_per_floor * self.surface) * 4 *
            (self.floor + 1) * self.ceiling_height
        ) * self.climate_zone * self.intermittence
        uen_per_metre2_2 = 1 + 0.1 * (
            self.ug * (self.dwelling_per_floor * self.surface * 2 +
                       math.sqrt(self.dwelling_per_floor * self.surface) * 4 *
                       (self.floor + 1) * self.ceiling_height) /
            (self.dwelling_per_floor * self.surface *
             (self.floor + 1) * self.ceiling_height / 2.5) - 1)
        uen_per_metre2_3 = self.dwelling_per_floor * self.surface * (
            self.floor + 1) * 1000
        self.uen_per_metre2 = uen_per_metre2_1 / uen_per_metre2_2 / uen_per_metre2_3
        return self.uen_per_metre2

    def UEN_per_Dwelling(self, p, s, m, w, r, f, df):
        self.uen_per_dwelling = self.uen_per_metre2 * self.surface
        return self.uen_per_dwelling

    def Call_Functions_Switchable(self, p, s, m, w, r, f, df):
        fonctions_list = [
            self.Housing_Categorizer(p, s, df),
            self.get_Housing_Type(p, s, df),
            self.get_Construction_Period(p, s, df),
            self.get_Size_Factor(p, s, df),
            self.get_Surface(p, s, df),
            self.get_Ceiling_Height(p, s, df),
            self.get_Share_of_Windows(p, s, df),
            self.get_Dwelling_per_Floor(p, s, df),
            self.get_Floor(p, s, df),
            self.get_Intermittence(p, s, df),
            self.get_Um(p, s, m, df),
            self.get_Uw(p, s, w, df),
            self.get_Ur(p, s, r, df),
            self.get_Uf(p, s, f, df),
            self.get_Climate_Zone(p, s, df),
            self.Surface_Composant_m(p, s, df),
            self.Surface_Composant_w(p, s, df),
            self.Surface_Composant_f(p, s, df),
            self.Surface_Composant_r(p, s, df),
            self.Surface_Composant_total(p, s, df),
            self.Ug(p, s, m, w, r, f, df),
            self.UEN_per_Metre2(p, s, m, w, r, f, df),
            self.UEN_per_Dwelling(p, s, m, w, r, f, df)
        ]
        return fonctions_list

    def GHC_Swithable(self, p, s, t_initial, t_switch, m, w, r, f, df):
        self.Call_Functions_Switchable(p, s, m, w, r, f, df)

        self.df_CD_CC_heating_system = Collective_Dwelling_CC_Heating_System(
            df_Heating_System)

        self.heating_system_to_switch = self.df_CD_CC_heating_system.get_Heating_System_Type(
            t_switch, df_Heating_System)

        self.initial_heating_system = self.df_CD_CC_heating_system.get_Heating_System_Type(
            t_initial, df_Heating_System)

        self.efficiency = self.df_CD_CC_heating_system.get_Efficiency(
            t_switch, df_Heating_System)
        self.a_investment = self.df_CD_CC_heating_system.get_A_Investment(
            t_switch, df_Heating_System)
        self.b_investment = self.df_CD_CC_heating_system.get_B_Investment(
            t_switch, df_Heating_System)
        self.maintenance = self.df_CD_CC_heating_system.get_Maintenance(
            t_switch, df_Heating_System)
        self.emission_factor = self.df_CD_CC_heating_system.get_Emission_Factor(
            t_switch, df_Heating_System)
        self.energy_price_v1 = self.df_CD_CC_heating_system.get_Energy_Price_v1(
            t_switch, df_Heating_System)

        self.same_switch_cost = self.df_CD_CC_heating_system.get_Same_Switch_Cost(
            t_switch, df_Heating_System)
        self.different_switch_cost = self.df_CD_CC_heating_system.get_Different_Switch_Cost(
            t_switch, df_Heating_System)
        self.delevery_type_cost = self.df_CD_CC_heating_system.get_Delevery_Type_Cost(
            t_switch, df_Heating_System)
        self.specific_gas_price = self.df_CD_CC_heating_system.get_Specific_Gas_Price(
            t_switch, df_Heating_System)

        CD_CC_initial_ghc_1 = (self.a_investment + self.b_investment *
                               self.uen_per_dwelling) / math.pow(
                                   (1 + self.discount_rate), 0)
        CD_CC_initial_ghc_2 = (self.a_investment + self.b_investment *
                               self.uen_per_dwelling) / math.pow(
                                   (1 + self.discount_rate), 20)
        CD_CC_initial_ghc_3 = self.uen_per_dwelling / self.efficiency
        CD_CC_initial_ghc_4 = 0
        CD_CC_initial_ghc_5 = self.maintenance

        if m == 0:
            self.CD_CC_cost_addition_m = self.surface_composant_m * 0
        else:
            M = m + 19
            self.CD_CC_cost_addition_m = self.surface_composant_m * float(
                df_Wall_Investment.iloc[:, 3].tolist()[M])

        if w == 0:
            self.CD_CC_cost_addition_w = self.surface_composant_w * 0
        else:
            W = w + 3
            self.CD_CC_cost_addition_w = self.surface_composant_w * float(
                df_Windows_Investment.iloc[:, 3].tolist()[W])

        if r == 0:
            self.CD_CC_cost_addition_r = self.surface_composant_r * 0
        else:
            R = r + 7
            self.CD_CC_cost_addition_r = self.surface_composant_r * float(
                df_Roof_Investment.iloc[:, 3].tolist()[R])

        if f == 0:
            self.CD_CC_cost_addition_f = self.surface_composant_f * 0
        else:
            F = f + 3
            self.CD_CC_cost_addition_f = self.surface_composant_f * float(
                df_Floor_Investment.iloc[:, 3].tolist()[F])

        self.CD_CC_insulation_cost = self.CD_CC_cost_addition_m + self.CD_CC_cost_addition_w + \
        self.CD_CC_cost_addition_r + self.CD_CC_cost_addition_f

        self.centralized_cost = self.delevery_type_cost * self.surface
        self.decentralized_cost = self.delevery_type_cost * self.surface
        self.control_cost = 0

        self.price_1 = self.energy_price_v1 * (
            1 +
            self.annual_variation) + self.emission_factor * self.carbon_price

        if self.heating_system_to_switch == self.initial_heating_system:

            self.specific_cost = 0

            CD_CC_initial_ghc_1 = (self.a_investment + self.b_investment *
                                   self.uen_per_dwelling) / math.pow(
                                       (1 + self.discount_rate), (19 - 0))
            CD_CC_initial_ghc_2 = (self.a_investment + self.b_investment *
                                   self.uen_per_dwelling) / math.pow(
                                       (1 + self.discount_rate), (39 - 0))
        else:
            if self.heating_system_to_switch == 'CC Wood Stove (WB)':

                if self.initial_heating_system in ['CC Gas Boiler (GB)', 'CC Condensing Gas Boiler (CGB)', 'CC Oil Boiler (OB)', \
                                           'CC Condensing Oil Boiler (COB)', 'CC Water Source Electric Heat Pump (EHP2)']:

                    self.control_cost = self.decentralized_cost

                    self.specific_cost = self.different_switch_cost

                elif self.initial_heating_system == 'CC Wood Pellet Boiler (WPB)':

                    self.control_cost = self.decentralized_cost

                    self.specific_cost = self.same_switch_cost

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch == 'CC Air Source Electric Heat Pump (EHP)':

                if self.initial_heating_system in ['CC Gas Boiler (GB)', 'CC Condensing Gas Boiler (CGB)', 'CC Oil Boiler (OB)', \
                                           'CC Condensing Oil Boiler (COB)', 'CC Water Source Electric Heat Pump (EHP2)', \
                                           'CC Wood Pellet Boiler (WPB)']:

                    self.control_cost = self.decentralized_cost

                    self.specific_cost = self.different_switch_cost

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch in [
                    'CC Gas Boiler (GB)', 'CC Condensing Gas Boiler (CGB)'
            ]:

                if self.initial_heating_system in [
                        'CC Gas Boiler (GB)', 'CC Condensing Gas Boiler (CGB)'
                ]:

                    self.specific_cost = self.same_switch_cost

                elif self.initial_heating_system in [
                        'CC Wood Stove (WB)',
                        'CC Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                elif self.initial_heating_system in [
                        'CC Oil Boiler (OB)', 'CC Condensing Oil Boiler (COB)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.price_1 = self.specific_gas_price * (
                        1 + self.annual_variation
                    ) + self.emission_factor * self.carbon_price

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch in [
                    'CC Oil Boiler (OB)', 'CC Condensing Oil Boiler (COB)'
            ]:

                if self.initial_heating_system in [
                        'CC Oil Boiler (OB)', 'CC Condensing Oil Boiler (COB)'
                ]:

                    self.specific_cost = self.same_switch_cost

                elif self.initial_heating_system in [
                        'CC Wood Stove (WB)',
                        'CC Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch == 'CC Wood Pellet Boiler (WPB)':

                if self.initial_heating_system == 'CC Wood Stove (WB)':

                    self.specific_cost = self.same_switch_cost

                    self.control_cost = self.centralized_cost

                elif self.initial_heating_system == 'CC Air Source Electric Heat Pump (EHP)':

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

            else:

                if self.initial_heating_system in [
                        'CC Wood Stove (WB)',
                        'CC Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

        for i in range(40):
            self.price = self.price_1 * math.pow(
                (1 + self.annual_variation), 1)
            CD_CC_initial_ghc_4 += self.price / math.pow(
                (1 + self.discount_rate), i + 1)

        self.CD_CC_ghc_swithable = CD_CC_initial_ghc_1 + CD_CC_initial_ghc_2 + \
        CD_CC_initial_ghc_3 * CD_CC_initial_ghc_4 + \
        CD_CC_initial_ghc_5 + self.CD_CC_insulation_cost + self.specific_cost + \
        self.control_cost
        return self.CD_CC_ghc_swithable
```

### `Individual_Dwelling(Housing)`

```python
class Individual_Dwelling(Housing):
    def __init__(self, df):
        super().__init__(df)
        self.df_Individual_Dwelling = self.df_housing.iloc[1:22, :]

    def Housing_Categorizer(self, p, s, df):
        if p == "<1968":
            if s == "Middle":
                self.housing = self.df_Individual_Dwelling.iloc[0].tolist()
            elif s == "Small":
                self.housing = self.df_Individual_Dwelling.iloc[1].tolist()
            elif s == "Large":
                self.housing = self.df_Individual_Dwelling.iloc[2].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1968-1973":
            if s == "Middle":
                self.housing = self.df_Individual_Dwelling.iloc[3].tolist()
            elif s == "Small":
                self.housing = self.df_Individual_Dwelling.iloc[4].tolist()
            elif s == "Large":
                self.housing = self.df_Individual_Dwelling.iloc[5].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1974-1981":
            if s == "Middle":
                self.housing = self.df_Individual_Dwelling.iloc[6].tolist()
            elif s == "Small":
                self.housing = self.df_Individual_Dwelling.iloc[7].tolist()
            elif s == "Large":
                self.housing = self.df_Individual_Dwelling.iloc[8].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1982-1989":
            if s == "Middle":
                self.housing = self.df_Individual_Dwelling.iloc[9].tolist()
            elif s == "Small":
                self.housing = self.df_Individual_Dwelling.iloc[10].tolist()
            elif s == "Large":
                self.housing = self.df_Individual_Dwelling.iloc[11].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "1990-2000":
            if s == "Middle":
                self.housing = self.df_Individual_Dwelling.iloc[12].tolist()
            elif s == "Small":
                self.housing = self.df_Individual_Dwelling.iloc[13].tolist()
            elif s == "Large":
                self.housing = self.df_Individual_Dwelling.iloc[14].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "2001-2005":
            if s == "Middle":
                self.housing = self.df_Individual_Dwelling.iloc[15].tolist()
            elif s == "Small":
                self.housing = self.df_Individual_Dwelling.iloc[16].tolist()
            elif s == "Large":
                self.housing = self.df_Individual_Dwelling.iloc[17].tolist()
            else:
                raise Exception('No matching size factor!')
        elif p == "2006-2013":
            if s == "Middle":
                self.housing = self.df_Individual_Dwelling.iloc[18].tolist()
            elif s == "Small":
                self.housing = self.df_Individual_Dwelling.iloc[19].tolist()
            elif s == "Large":
                self.housing = self.df_Individual_Dwelling.iloc[20].tolist()
            else:
                raise Exception('No matching size factor!')
        else:
            raise Exception('No matching construction period!')
        return self.housing

    def Surface_Composant_m(self, p, s, df):
        self.surface_composant_m = (1 - self.share_of_windows) * math.sqrt(
            self.surface /
            (self.floor + 1)) * 4 * (self.floor + 1) * self.ceiling_height
        return self.surface_composant_m

    def Surface_Composant_w(self, p, s, df):
        self.surface_composant_w = self.share_of_windows * math.sqrt(
            self.surface /
            (self.floor + 1)) * 4 * (self.floor + 1) * self.ceiling_height
        return self.surface_composant_w

    def Ug(self, p, s, df):
        ug_1 = (
            (1 - self.share_of_windows) * self.um +
            self.share_of_windows * self.uw) * math.sqrt(
                self.surface / (self.floor + 1)) * 4 * self.ceiling_height * (
                    self.floor + 1) + (self.ur + self.uf) * self.surface / (
                        self.floor + 1)

        ug_2 = math.sqrt(
            self.surface / (self.floor + 1)) * 4 * self.ceiling_height * (
                self.floor + 1) + 2 * self.surface / (self.floor + 1)
        self.ug = ug_1 / ug_2
        return self.ug

    def UEN_per_Metre2(self, p, s, df):
        uen_per_metre2_1 = math.sqrt(
            self.surface / (self.floor + 1)) * 4 * self.ceiling_height * (
                self.floor + 1) + 2 * self.surface / (self.floor + 1)
        uen_per_metre2_2 = self.ug * self.intermittence * self.climate_zone / (
            self.surface * math.pow(10, 3))
        uen_per_metre2_3 = 1 + 0.1 * (
            (math.sqrt(self.surface /
                       (self.floor + 1)) * 4 * self.ceiling_height *
             (self.floor + 1) + 2 * self.surface /
             (self.floor + 1)) * self.ug /
            (self.surface * self.ceiling_height / 2.5) - 1)
        self.uen_per_metre2 = uen_per_metre2_1 * uen_per_metre2_2 / uen_per_metre2_3
        return self.uen_per_metre2

    def Call_Functions(self, p, s, t_switch, df):
        fonctions_list = [
            self.Housing_Categorizer(p, s, df),
            self.get_Housing_Type(p, s, df),
            self.get_Construction_Period(p, s, df),
            self.get_Size_Factor(p, s, df),
            self.get_Surface(p, s, df),
            self.get_Ceiling_Height(p, s, df),
            self.get_Share_of_Windows(p, s, df),
            self.get_Dwelling_per_Floor(p, s, df),
            self.get_Floor(p, s, df),
            self.get_Intermittence(p, s, df),
            self.get_Um(p, s, df),
            self.get_Uw(p, s, df),
            self.get_Ur(p, s, df),
            self.get_Uf(p, s, df),
            self.get_Climate_Zone(p, s, df),
            self.Surface_Composant_m(p, s, df),
            self.Surface_Composant_w(p, s, df),
            self.Surface_Composant_f(p, s, df),
            self.Surface_Composant_r(p, s, df),
            self.Surface_Composant_total(p, s, df),
            self.Ug(p, s, df),
            self.UEN_per_Metre2(p, s, df),
            self.UEN_per_Dwelling(p, s, df),
        ]
        return fonctions_list

    def Initial_GHC(self, p, s, t_switch, df):
        self.Call_Functions(p, s, t_switch, df)
        self.df_ID_heating_system = Individual_Dwelling_Heating_System(
            df_Heating_System)

        self.heating_system_to_switch = self.df_ID_heating_system.get_Heating_System_Type(
            t_switch, df_Heating_System)
        self.efficiency = self.df_ID_heating_system.get_Efficiency(
            t_switch, df_Heating_System)
        self.a_investment = self.df_ID_heating_system.get_A_Investment(
            t_switch, df_Heating_System)
        self.b_investment = self.df_ID_heating_system.get_B_Investment(
            t_switch, df_Heating_System)
        self.maintenance = self.df_ID_heating_system.get_Maintenance(
            t_switch, df_Heating_System)
        self.emission_factor = self.df_ID_heating_system.get_Emission_Factor(
            t_switch, df_Heating_System)
        self.energy_price_v1 = self.df_ID_heating_system.get_Energy_Price_v1(
            t_switch, df_Heating_System)

        initial_ghc_1 = (self.a_investment +
                         self.b_investment * self.uen_per_dwelling) / math.pow(
                             (1 + self.discount_rate), (19 - 0))
        initial_ghc_2 = (self.a_investment +
                         self.b_investment * self.uen_per_dwelling) / math.pow(
                             (1 + self.discount_rate), (39 - 0))
        initial_ghc_3 = self.uen_per_dwelling / self.efficiency
        initial_ghc_4 = 0
        self.price_1 = self.energy_price_v1 * (
            1 +
            self.annual_variation) + self.emission_factor * self.carbon_price
        for i in range(40):
            self.price = self.price_1 * math.pow(
                (1 + self.annual_variation), 1)
            initial_ghc_4 += self.price / math.pow(
                (1 + self.discount_rate), i + 1)
        initial_ghc_5 = self.maintenance
        self.insulation_cost = 0
        self.initial_ghc = initial_ghc_1 + initial_ghc_2 + \
        initial_ghc_3 * initial_ghc_4 + initial_ghc_5 + \
        self.insulation_cost
        return self.initial_ghc
```

```python
class Individual_Dwelling_per_Year_of_HS(
        Individual_Dwelling_Switchable_Insulation):
    def __init__(self, df):
        super().__init__(df)

    def Call_Functions_Switchable(self, p, s, m, w, r, f, df):
        fonctions_list = [
            self.Housing_Categorizer(p, s, df),
            self.get_Housing_Type(p, s, df),
            self.get_Construction_Period(p, s, df),
            self.get_Size_Factor(p, s, df),
            self.get_Surface(p, s, df),
            self.get_Ceiling_Height(p, s, df),
            self.get_Share_of_Windows(p, s, df),
            self.get_Dwelling_per_Floor(p, s, df),
            self.get_Floor(p, s, df),
            self.get_Intermittence(p, s, df),
            self.get_Um(p, s, m, df),
            self.get_Uw(p, s, w, df),
            self.get_Ur(p, s, r, df),
            self.get_Uf(p, s, f, df),
            self.get_Climate_Zone(p, s, df),
            self.Surface_Composant_m(p, s, df),
            self.Surface_Composant_w(p, s, df),
            self.Surface_Composant_f(p, s, df),
            self.Surface_Composant_r(p, s, df),
            self.Surface_Composant_total(p, s, df),
            self.Ug(p, s, m, w, r, f, df),
            self.UEN_per_Metre2(p, s, m, w, r, f, df),
            self.UEN_per_Dwelling(p, s, m, w, r, f, df)
        ]
        return fonctions_list

    def GHC_Swithable_per_Year(self, p, s, t_initial, t_switch, m, w, r, f, df,
                               year):
        self.Call_Functions_Switchable(p, s, m, w, r, f, df)

        self.df_ID_heating_system = Individual_Dwelling_Heating_System(
            df_Heating_System)

        self.heating_system_to_switch = self.df_ID_heating_system.get_Heating_System_Type(
            t_switch, df_Heating_System)

        self.initial_heating_system = self.df_ID_heating_system.get_Heating_System_Type(
            t_initial, df_Heating_System)

        self.efficiency = self.df_ID_heating_system.get_Efficiency(
            t_switch, df_Heating_System)
        self.a_investment = self.df_ID_heating_system.get_A_Investment(
            t_switch, df_Heating_System)
        self.b_investment = self.df_ID_heating_system.get_B_Investment(
            t_switch, df_Heating_System)
        self.maintenance = self.df_ID_heating_system.get_Maintenance(
            t_switch, df_Heating_System)
        self.emission_factor = self.df_ID_heating_system.get_Emission_Factor(
            t_switch, df_Heating_System)
        self.energy_price_v1 = self.df_ID_heating_system.get_Energy_Price_v1(
            t_switch, df_Heating_System)

        self.same_switch_cost = self.df_ID_heating_system.get_Same_Switch_Cost(
            t_switch, df_Heating_System)
        self.different_switch_cost = self.df_ID_heating_system.get_Different_Switch_Cost(
            t_switch, df_Heating_System)
        self.delevery_type_cost = self.df_ID_heating_system.get_Delevery_Type_Cost(
            t_switch, df_Heating_System)
        self.specific_gas_price = self.df_ID_heating_system.get_Specific_Gas_Price(
            t_switch, df_Heating_System)

        ID_initial_ghc_1 = (self.a_investment + self.b_investment *
                            self.uen_per_dwelling) / math.pow(
                                (1 + self.discount_rate), 0)
        ID_initial_ghc_2 = (self.a_investment + self.b_investment *
                            self.uen_per_dwelling) / math.pow(
                                (1 + self.discount_rate), 20)
        ID_initial_ghc_3 = self.uen_per_dwelling / self.efficiency
        ID_initial_ghc_4 = 0
        ID_initial_ghc_5 = self.maintenance

        if m == 0:
            self.ID_cost_addition_m = self.surface_composant_m * 0
        else:
            M = m + 0
            self.ID_cost_addition_m = self.surface_composant_m * float(
                df_Wall_Investment.iloc[:, 3].tolist()[M])

        if w == 0:
            self.ID_cost_addition_w = self.surface_composant_w * 0
        else:
            W = w + 0
            self.ID_cost_addition_w = self.surface_composant_w * float(
                df_Windows_Investment.iloc[:, 3].tolist()[W])

        if r == 0:
            self.ID_cost_addition_r = self.surface_composant_r * 0
        else:
            R = r + 0
            self.ID_cost_addition_r = self.surface_composant_r * float(
                df_Roof_Investment.iloc[:, 3].tolist()[R])

        if f == 0:
            self.ID_cost_addition_f = self.surface_composant_f * 0
        else:
            F = f + 0
            self.ID_cost_addition_f = self.surface_composant_f * float(
                df_Floor_Investment.iloc[:, 3].tolist()[F])

        self.ID_insulation_cost = self.ID_cost_addition_m + self.ID_cost_addition_w + \
        self.ID_cost_addition_r + self.ID_cost_addition_f

        self.centralized_cost = self.delevery_type_cost * self.surface
        self.decentralized_cost = self.delevery_type_cost * self.surface
        self.control_cost = 0

        self.price_1 = self.energy_price_v1 * (
            1 +
            self.annual_variation) + self.emission_factor * self.carbon_price

        if self.heating_system_to_switch == self.initial_heating_system:

            self.specific_cost = 0

            ID_initial_ghc_1 = (self.a_investment + self.b_investment *
                                self.uen_per_dwelling) / math.pow(
                                    (1 + self.discount_rate), (19 - year))
            ID_initial_ghc_2 = (self.a_investment + self.b_investment *
                                self.uen_per_dwelling) / math.pow(
                                    (1 + self.discount_rate), (39 - year))
        else:
            if self.heating_system_to_switch in [
                    'Electric Heater (ER)', 'Wood Stove (WB)',
                    'Air Source Electric Heat Pump (EHP)'
            ]:
                if self.initial_heating_system in ['Gas Boiler (GB)', 'Condensing Gas Boiler (CGB)', 'Oil Boiler (OB)', \
                                           'Condensing Oil Boiler (COB)', 'Water Source Electric Heat Pump (EHP2)', \
                                           'Wood Pellet Boiler (WPB)']:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.decentralized_cost

                else:

                    self.specific_cost = self.same_switch_cost

            elif self.heating_system_to_switch in [
                    'Gas Boiler (GB)', 'Condensing Gas Boiler (CGB)'
            ]:

                if self.initial_heating_system in [
                        'Gas Boiler (GB)', 'Condensing Gas Boiler (CGB)'
                ]:

                    self.specific_cost = self.same_switch_cost

                elif self.initial_heating_system in [
                        'Electric Heater (ER)', 'Wood Stove (WB)',
                        'Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                elif self.initial_heating_system in [
                        'Oil Boiler (OB)', 'Condensing Oil Boiler (COB)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.price_1 = self.specific_gas_price * (
                        1 + self.annual_variation
                    ) + self.emission_factor * self.carbon_price

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch in [
                    'Oil Boiler (OB)', 'Condensing Oil Boiler (COB)'
            ]:

                if self.initial_heating_system in [
                        'Oil Boiler (OB)', 'Condensing Oil Boiler (COB)'
                ]:

                    self.specific_cost = self.same_switch_cost

                elif self.initial_heating_system in [
                        'Electric Heater (ER)', 'Wood Stove (WB)',
                        'Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

            elif self.heating_system_to_switch == 'Wood Pellet Boiler (WPB)':

                if self.initial_heating_system == 'Wood Stove (WB)':

                    self.specific_cost = self.same_switch_cost

                    self.control_cost = self.centralized_cost

                elif self.initial_heating_system in [
                        'Electric Heater (ER)',
                        'Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

            else:

                if self.initial_heating_system in [
                        'Electric Heater (ER)', 'Wood Stove (WB)',
                        'Air Source Electric Heat Pump (EHP)'
                ]:

                    self.specific_cost = self.different_switch_cost

                    self.control_cost = self.centralized_cost

                else:

                    self.specific_cost = self.different_switch_cost

        for i in range(40):
            self.price = self.price_1 * math.pow(
                (1 + self.annual_variation), 1)
            ID_initial_ghc_4 += self.price / math.pow(
                (1 + self.discount_rate), i + 1)

        self.ID_ghc_swithable = ID_initial_ghc_1 + ID_initial_ghc_2 + \
        ID_initial_ghc_3 * ID_initial_ghc_4 + \
        ID_initial_ghc_5 + self.ID_insulation_cost + self.specific_cost + \
        self.control_cost
        return self.ID_ghc_swithable
```

## `Heating_System_Switching_Rules`

```python
class Heating_System_Switching_Rules:
    def __init__(self, df):
        self.df = df

    def Finding_Minimum_to_Insulate(self, df):

        self.series_min = self.df.iloc[self.df['Initial HS'].idxmin(), :]

        return self.series_min

    def Finding_Minimum_to_Queue(self, df):

        self.df = self.df.iloc[0:9, :]

        self.series_min = self.df.iloc[self.df['Initial HS'].idxmin(), :]

        return self.series_min

    def Minimum_m_w_r_f(self, df):

        self.m_min = self.series_min.tolist()[3]
        self.w_min = self.series_min.tolist()[4]
        self.r_min = self.series_min.tolist()[5]
        self.f_min = self.series_min.tolist()[6]

        for letter in ['M', 'W', 'F', 'R']:
            self.m_min = self.m_min.replace(letter, '')
            self.w_min = self.w_min.replace(letter, '')
            self.r_min = self.r_min.replace(letter, '')
            self.f_min = self.f_min.replace(letter, '')

        self.m_min = int(self.m_min)
        self.w_min = int(self.w_min)
        self.r_min = int(self.r_min)
        self.f_min = int(self.f_min)

        return self.m_min, self.w_min, self.r_min, self.f_min

    def Changing_Heating_System(self, df, t_initial, year):
        self.year = year
        self.t_initial = t_initial
        if self.series_min['HS to Switch'] == self.t_initial:
            if self.year < 19:
                self.year += 1
            else:
                self.year = 0
        else:
            self.year = 0
            self.t_initial = self.series_min['HS to Switch']
        return self.t_initial, self.year
```
