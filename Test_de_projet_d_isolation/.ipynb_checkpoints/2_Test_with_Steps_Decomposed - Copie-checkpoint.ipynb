# Combination of all the necessary data

```python
class All_Types:
    def __init__(self, *args):
        self.df_1 = df_Climate_Zone
        self.df_2 = df_Housing_Characteristics
        self.df_3 = df_Headcount
        self.df_4 = df_Heating_System
        self.df_5 = df_Economic_Hypothesis
        self.df_6 = df_Wall_Investment
        self.df_7 = df_Window_Investment
        self.df_8 = df_Roof_Investment
        self.df_9 = df_Floor_Investment
```

```python
class Type_CID(All_Types):
    def init(self, *args):
        super().init(*args)

    def List_Itertools(self):
        self.list_itertools = []
        list_dwelling_type = ['CID']
        list_climate_type = self.df_1.iloc[:, 0].tolist()
        list_construction_period = self.df_2.iloc[:21, 1].tolist()[::3]
        list_size_factor = self.df_2.iloc[:3, 2].tolist()
        list_heating_system = [
            'ER', 'GB', 'CGB', 'OB', 'COB', 'EHP2', 'WB', 'WPB', 'EHP'
        ]
        list_initial_age = self.df_3.iloc[:20, 5].tolist()
        self.list_itertools.append(list_dwelling_type)
        self.list_itertools.append(list_climate_type)
        self.list_itertools.append(list_construction_period)
        self.list_itertools.append(list_size_factor)
        self.list_itertools.append(list_heating_system)
        self.list_itertools.append(list_initial_age)
        return self.list_itertools

    def List_to_DF(self):
        self.List_Itertools()
        itertools_product = list(itertools.product(*self.list_itertools))
        self.df = pd.DataFrame(itertools_product,
                               columns=[
                                   'Dwelling Type', 'Climate Type',
                                   'Construction Period', 'Size Factor',
                                   'Heating System', 'Initial Age'
                               ])
        return self.df

    def DFs_Merge(self):
        self.List_to_DF()
        self.df = self.df.merge(self.df_1, how='left', on=['Climate Type'])
        self.df = self.df.merge(self.df_2.iloc[:21, 1:9],
                                how='left',
                                on=['Construction Period', 'Size Factor'])
        self.df = self.df.merge(self.df_4.iloc[:9, :],
                                how='left',
                                on=['Heating System'])
        return self.df

    def Cols_Add(self):
        self.DFs_Merge()
        self.df[self.df_5.columns[0]] = self.df_5.iloc[0, 0]
        self.df[self.df_5.columns[1]] = self.df_5.iloc[0, 1]
        self.df[self.df_5.columns[2]] = self.df_5.iloc[0, 2]
        self.df[self.df_5.columns[3]] = self.df_5.iloc[0, 3]
        return self.df

    def Complete_DF_without_U(self):
        self.Cols_Add()
        self.df = pd.concat([self.df, self.df_3.iloc[0:11340, 6]],
                            axis=1,
                            join='outer')
        return self.df

    def Cols_Reorder(self):
        self.Complete_DF_without_U()
        cols = self.df.columns.tolist()
        reordered_cols = cols[:2] + cols[6:7] + cols[2:4] + cols[7:13] + \
        cols[4:5] + cols[13:27] + cols[5:6] + cols[27:28]
        self.df = self.df[reordered_cols]
        return self.df

    def DF_with_U_Factors(self):
        self.Cols_of_Initial_U()
        self.df = self.df.merge(
            self.df_2.loc[:20, ['Construction Period', 'Size Factor', 'Um0']],
            how='left',
            on=['Construction Period', 'Size Factor'])
        for m in range(18):
            self.df['Um' + str(m + 1)] = self.df_6.iloc[m, 2]
        self.df = self.df.merge(
            self.df_2.loc[:20, ['Construction Period', 'Size Factor', 'Uw0']],
            how='left',
            on=['Construction Period', 'Size Factor'])
        for w in range(3):
            self.df['Uw' + str(w + 1)] = self.df_7.iloc[w, 2]
        self.df = self.df.merge(
            self.df_2.loc[:20, ['Construction Period', 'Size Factor', 'Ur0']],
            how='left',
            on=['Construction Period', 'Size Factor'])
        for r in range(3):
            self.df['Ur' + str(r + 1)] = self.df_8.iloc[r, 2]
        self.df = self.df.merge(
            self.df_2.loc[:20, ['Construction Period', 'Size Factor', 'Uf0']],
            how='left',
            on=['Construction Period', 'Size Factor'])
        for f in range(5):
            self.df['Uf' + str(f + 1)] = self.df_9.iloc[f, 2]
        return self.df

    def DF_with_U_Prices(self):
        self.Cols_of_Initial_U()
        self.df['Um0'] = 0
        for m in range(18):
            self.df['Um' + str(m + 1)] = self.df_6.iloc[m, 3]

        self.df['Uw0'] = 0
        for w in range(3):
            self.df['Uw' + str(w + 1)] = self.df_7.iloc[w, 3]

        self.df['Ur0'] = 0
        for r in range(3):
            self.df['Ur' + str(r + 1)] = self.df_8.iloc[r, 3]

        self.df['Uf0'] = 0
        for f in range(5):
            self.df['Uf' + str(f + 1)] = self.df_9.iloc[f, 3]
        return self.df
```

```python
class First_Year_CID(Type_CID):
    def init(self, *args):
        super().init(*args)

    def Cols_of_Initial_U(self):
        self.Cols_Reorder()
        self.df['Initial Um'] = 0
        self.df['Initial Uw'] = 0
        self.df['Initial Ur'] = 0
        self.df['Initial Uf'] = 0
        return self.df
```

```python
First_Year_CID = First_Year_CID()
df_with_U_Factors = First_Year_CID.DF_with_U_Factors()
df_with_U_Prices = First_Year_CID.DF_with_U_Prices()
```

```python
class Calculable_Cols:
    def __init__(self, df):
        self.df = df
```

```python
class Cols_for_Collective_Dwelling(Calculable_Cols):
    def init(self, df):
        super().init(df)

    def Cols_Rename(self):
        self.df.rename(columns={
            name: name.replace(' ', '_')
            for name in self.df.columns if name.count(' ') > 0
        },
                       inplace=1)
        return self.df

    def Col_Surface_m(self):
        self.Cols_Rename()
        A = np.subtract(1, self.df.Share_of_Windows)
        B = np.sqrt(np.multiply(self.df.Surface, self.df.Dwelling_per_Floor))
        C = np.add(self.df.Floors, 1)
        D = np.multiply(np.multiply(A, B), 4)
        E = np.multiply(self.df.Ceiling_Height, C)
        self.df['Surface_m'] = np.multiply(D, E)
        return self.df

    def Col_Surface_w(self):
        self.Cols_Rename()
        A = np.sqrt(np.multiply(self.df.Surface, self.df.Dwelling_per_Floor))
        B = np.add(self.df.Floors, 1)
        C = np.multiply(np.multiply(self.df.Share_of_Windows, A), 4)
        D = np.multiply(self.df.Ceiling_Height, B)
        self.df['Surface_w'] = np.multiply(C, D)
        return self.df

    def Col_Surface_r(self):
        self.Cols_Rename()
        self.df['Surface_r'] = np.multiply(self.df.Surface,
                                           self.df.Dwelling_per_Floor)
        return self.df

    def Weighted_U_Factors(self):
        self.Col_Surface_m()
        self.Col_Surface_w()
        self.Col_Surface_r()
        self.df.loc[:, 'Um0':'Um18'] = self.df.loc[:, 'Um0':'Um18'].apply(
            lambda x: np.multiply(x, self.df.Surface_m))
        self.df.loc[:, 'Uw0':'Uw3'] = self.df.loc[:, 'Uw0':'Uw3'].apply(
            lambda x: np.multiply(x, self.df.Surface_w))
        self.df.loc[:, 'Ur0':'Ur3'] = self.df.loc[:, 'Ur0':'Ur3'].apply(
            lambda x: np.multiply(x, self.df.Surface_r))
        self.df.loc[:, 'Uf0':'Uf5'] = self.df.loc[:, 'Uf0':'Uf5'].apply(
            lambda x: np.multiply(x, self.df.Surface_r))
        return self.df

    #######
    def Col_Dwelling_total(self):
        self.Cols_Rename()
        A = np.add(self.df.Floors, 1)
        self.df['Dwelling_total'] = np.multiply(self.df.Dwelling_per_Floor,\
                                                A)
        return self.df
```

```python
class Calculable_DFs:
    def __init__(self, df):
        self.df = df
```

```python
class DFs_for_Collective_Dwelling(Calculable_DFs):
    def init(self, df):
        super().init(df)

    def Cols_to_Swith_Add(self):
        cols_list = []
        for m in range(19):
            for w in range(4):
                for r in range(4):
                    for f in range(6):
                        for to_swith in [
                                'ER', 'GB', 'CGB', 'OB', 'COB', 'EHP2', 'WB',
                                'WPB', 'EHP'
                        ]:
                            col_name = str(m) + '_' + str(w) + '_' + str(
                                r) + '_' + str(f) + '_' + to_swith
                            cols_list.append(col_name)
        self.df = pd.concat(
            [self.df,
             pd.DataFrame(0, index=self.df.index, columns=cols_list)],
            axis=1)
        return self.df

    def Combine_Weighted_U_Factors(self):
        self.Cols_to_Swith_Add()
        selected_cols = self.df.columns.tolist()[-16416:]
        self.df[selected_cols] = self.df[selected_cols].apply(lambda x: np.add(
            np.add(self.df['Um' + x.name.split('_')[0]], self.df[
                'Uw' + x.name.split('_')[1]]),
            np.add(self.df['Ur' + x.name.split('_')[2]], self.df[
                'Uf' + x.name.split('_')[3]])))
        return self.df
```
